```python
path="/home/zuznanna/IdeaProjects/VaccineDistribution/out"
```

```python
total_wages=[]
total_ill=[]
for cost in range(1,6):
    files = ["n"+str(cost)+".txt"+str(i) for i in range(0,10)]
    print(files)
    wages_sum_array = []
    number_of_residents_array = []
    number_of_ill_array = []
    number_of_vacinated_array = []

    for fileName in files:
        f = open(path+"/{}".format(fileName), "r")
        splited = f.readline().rstrip("\n").split(";")
        wages_sum_array.append(int(splited[0]))
        number_of_residents_array.append(int(splited[1]))
        number_of_ill_array.append(int(splited[2]))
        number_of_vacinated_array.append(int(splited[3]))
    
    total_wages.append(sum(wages_sum_array)/len(wages_sum_array))
    total_ill.append(sum(number_of_ill_array)/len(number_of_ill_array))
```

```python
x = [i for i in range(1, len(total_wages)+1)]

my_xticks = ['1:0','0.75:0.25','0.5:0.5','0.25:0.75','1:0']


fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
# ax1.scatter(x, total_wages, s=500 ,label="Suma ścieżek")
ax1.plot(x, total_wages, label="Suma ścieżek")
ax2.plot(x, total_ill,c="r", label="Suma chorych")
ax1.yaxis.set_label_text('Suma ścieżek')
ax2.yaxis.set_label_text('Suma chorych')
ax1.xaxis.set_label_text('Stosunek parametru wagi funkcji celu: waga_krawędzi:waga_chorych')
plt.xticks(x, my_xticks)
# fig.xlabel('parametr wagi funkcji celu')
# fig.ylabel('suma wag')
plt.title("Wykres dla grafu o 100 nodach, uśredniony po 10 wykonaniach z różnymi ścieżkami początkowymi", y=1.1)
fig.legend(bbox_to_anchor=(1.3, 0.2))
plt.show()
```
